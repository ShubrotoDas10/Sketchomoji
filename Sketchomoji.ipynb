```python
# Step 1: Import Required Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
from PIL import Image
import io
```

```python
# Step 2: Upload an Image
uploaded = files.upload()
for fn in uploaded.keys():
    image_path = fn
    print(f"Uploaded file: {image_path}")
```

```python
# Step 3: Read & Show the Original Image
image = cv2.imread(image_path)
image = cv2.resize(image, (300, 300))  # Resize for simplicity
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Step 4: Apply Canny Edge Detection
edges = cv2.Canny(gray, threshold1=50, threshold2=150)

plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detected Borders")
plt.axis("off")
plt.show()
```

```python
# Step 5: Define the ASCII characters to use (dense to light)
ascii_chars = ['@', '#', '8', '&', '*', '=', '-', ':', '.', ' ']  # You can customize this
# You can also try symbols like: ascii_chars = ['<', '|', '_', ',', '.', ' ']

#  Step 6: Resize edges image for ASCII output (smaller height for terminal aspect ratio)
ascii_width = 100  # Reduce to make output manageable
height, width = edges.shape
aspect_ratio = height / width
ascii_height = int(ascii_width * aspect_ratio * 0.55)  # 0.55 fixes aspect distortion

resized = cv2.resize(edges, (ascii_width, ascii_height))
```

```python
# Step 6: Assume `edges` is your Canny output (from previous step)

# Define your preferred ASCII symbols to use for edge pixels
symbols = ['<', '|', '_', '/', '\\', ',']

# Randomly assign one of the symbols to each edge pixel
import random

ascii_output = ""

for row in edges:
    for pixel in row:
        if pixel == 255:
            # It's an edge pixel
            ascii_output += random.choice(symbols)
        else:
            # Not an edge, leave as blank space
            ascii_output += " "
    ascii_output += "\n"

#  Print result
print(ascii_output)
```

```python
# Save to file
with open("ascii_edges_only.txt", "w") as f:
    f.write(ascii_output)

# Download (Colab)
from google.colab import files
files.download("ascii_edges_only.txt")
```
